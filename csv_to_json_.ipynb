```python
import csv
import json
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
# read in and merge data
df_clin = pd.read_csv('./prad_msk_stopsack_2021_clinical_data.tsv', sep='\t')
df_cna = pd.read_csv('./data_cna.txt', sep='\t')

df_cna = df_cna.set_index('Hugo_Symbol').T.reset_index().rename(columns={'index':'Sample_ID'})
df = pd.merge(df_clin, df_cna, how='left', left_on='Sample ID', right_on='Sample_ID').drop(columns='Sample_ID')
```
